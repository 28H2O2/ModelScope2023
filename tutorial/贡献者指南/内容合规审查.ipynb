本篇文章介绍ModelScope社区的内容合规审查政策。

ModelScope平台鼓励开发者分享优质、好用的模型，提升算法开发效率，加速模型研发创新。在使用过程中，ModelScope社区希望用户承担内容分享、内容传播应尽的责任，不得分享违反中国法律、以及社会公德的信息内容，详细情景参考[用户协议与隐私政策 ](../用户协议/用户协议与隐私政策.ipynb)。在法律允许的范围内，我们对您创建或上传的内容任何公开显示或滥用不承担责任，但我们有权删除我们或有权的监管机构认为违反了任何法律或平台协议或政策的内容，以保障健康友好的社区环境。
# 审查标准
我们将在文件上传、信息填写、Demo服务等环节进行内容合规审查，审查标准如下：

- 是否含有涉政内容
- 是否含有色情内容
- 是否含有暴恐内容
- 是否含有违禁内容
- 是否含有广告及无意义灌水等信息
# 审查范围
若您上传的**文本、图片、音频、视频、文件**，以及**文件附带的提交信息（例如commit信息）**内含有以上违规内容，我们将有权删除违规信息及文件，并告知您处理信息。审查范围包括但不仅限于用户上传的数据集文件、模型文件以及模型Demo的输入。
# 审查流程
您通过网页或Git命令上传模型文件后，将在网页端模型及数据集文件列表内看到以下提示，告知文件正在审核中，审核时间一般在3-5分钟内：
![image.png](./resources/1660120625927-b8e757b9-49e0-4cb9-a34d-98a90f460afd.png)
若上传的模型及数据集文件违规，您将在网页端模型及数据集文件列表内查看到如下提示：
![image.png](./resources/1660120578739-9b969091-413d-43f0-9ffa-91f84cd10e4c.png)

若Demo输入违规，您将在输入框下方收到如下提示：

![image.png](./resources/1659504297539-9ad7cab3-86b0-423b-a51f-4b44e394d6fa.png)
若审查通过，您上传的文件将转为对外可访问状态。
# 如何解决
若您上传的违规或疑似违规文件被删除，您可自行查验文件内容，修改后重新上传。
