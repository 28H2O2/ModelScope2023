本篇文章向您详细介绍数据集的文件维护功能及注意事项，数据集创建请参考[数据集介绍](./数据集介绍.ipynb#创建数据集)中创建数据集部分。
# 核心概念

- 【元数据文件】元数据是描述数据的数据，主要包含描述数据的内容、属性信息、存储位置、使用方式等。根据托管方式不同，有两种形态：
   - 托管到公开Host：
      - 同名Python文件（由系统自动创建），用户需要完善Python文件，完成对数据的解析。格式详见[数据集文件规范](./数据集文件规范.ipynb#托管到公开Host数据集文件规范)。
   - 托管到ModelScope：
      - CSV文件：用于描述数据结构或直接存储数据。如涉及音频/视频/图片等形态文件，CSV文件存储数据相对路径；如是文本，CSV文件可以直接存文本信息；格式详见[数据集文件规范](./数据集文件规范.ipynb#CSV文件)。
      - 同名JSON文件：用于描述数据集、子数据集、用途之间的关系。格式详见[数据集文件规范](./数据集文件规范.ipynb#同名JSON文件)。
- 【数据文件】实际的数据文件，根据托管方式不同，有两种形态：
   - 托管到公开Host：
      - 一般是一个或者多个压缩文件（如ZIP等格式），由【元数据文件】->【同名Python文件】->【_HOMEPAGE全局变量】指定。
   - 托管到ModelScope：
      - 一个或者多个压缩文件（如ZIP等格式），在【元数据文件】->【CSV文件】中引用。
      - 多文件：扫描并上传指定目录内的多个数据文件（非压缩文件）。仅Dataset SDK支持该形式数据文件，Web端不支持，在【元数据文件】->【CSV文件】中引用。
# 数据集维护
![image.png](./resources/1659854600601-8780585c-f957-4bb9-a72c-55e581ce7521.png)
数据集介绍内容由元数据文件中的README.ipynb文件解析生成。为方便用户使用，建议包含数据集描述、简介、数据格式、数据集加载方式、数据集版权、引用方式等内容。
为方便维护README.ipynb文件，系统提供了对应文件[生成器小工具](https://www.modelscope.cn/#/readmeformatter)，您可以在手工选择后，直接粘贴自动生成的内容到对应编辑窗口，快速生成数据集描述内容。
# 数据预览
![image.png](./resources/1659854718253-6e385ea5-61cf-4c25-a500-4d32e2c1b57e.png)
可以预览数据集中的数据文件，支持按子数据集、用途，预览前100条数据。预览内容由系统自动生成。
如未能自动生成预览内容，可能的原因有：

- 数据集正在解析中，尚未生成预览数据；
- 未按规范要求，提供正确的数据集信息，[数据集文件规范](./数据集文件规范.ipynb#README.ipynb文件)；
- 由于网络受限原因，导致系统无法访问托管于公开Host的数据集；
- 托管到ModelScope的数据集，zip包大于等于5GB时，不支持预览功能；
# 数据文件
根据数据集创建时托管方式不同，分为以下两种情况：
## 托管到公开Host

- 如采用官方推荐模式，可直接编辑同名Python文件，对_HOMEPAGE全局变量进行赋值，即表示数据文件的实际托管位置，例：_HOMEPAGE = "https://github.com/zalandoresearch/fashion-mnist"。
- 如使用自行提供的数据解析文件，请根据实际情况对数据引用链接进行配置。
- 目前我们支持引用的Python三方库：pyarrow, datasets, Pillow, soundfile, torchaudio, oss2，如果引入了其他三方库，可能会导致解析失败，如有问题，请联系我们。
## 托管到ModelScope

- 在【元数据文件】标签页，上传CSV文件。
- 在【数据文件】标签页，上传ZIP文件。
- 在【元数据文件】标签页，同名JSON文件，维护子数据集、用途、CSV文件、ZIP文件对应关系，可通过页面进行编辑。
# 删除数据集
您只能对自己拥有或管理的数据集进行管理设置操作。请前往至需要删除的数据集卡片，并点击“设置”页面，在这里您可以编辑数据集的中文名称、描述，修改数据集的开源许可证协议或数据集的可见权限。您也可以在此页面删除数据集。

![image.png](./resources/1660923430468-b8a09d0c-f9d5-4793-9361-025cb81bdbcc.png)

需要注意的是，若您需要删除数据集，需要进行二次验证。且该删除操作不可撤销，数据集中的所有文件将被永久性删除，请谨慎操作！
