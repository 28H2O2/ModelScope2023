本文整理了ModelScope Notebook使用过程中遇到的常见问题，希望能够通过此文档为您解答使用过程中的疑惑。

**Q1：打开实例时间较长怎么办？**
由于机器资源需要排队，且每个实例需要预装ModelScope官方镜像，请您耐心等待，通常CPU实例打开时间在5分钟以内，GPU在排队情况下打开时间在10分钟以内。实例启动过程中，您可以打开新标签页查看其他模型介绍。若等待时间仍然较长，请您加入**ModelScope开发者联盟钉钉交流群**联系管理员进行解决。

**Q2：实例单次启动时间超过了4小时，实例关闭了怎么办？**
实例单次启动时间超过了4小时后，您可以在Notebook内保存已编写的代码文件，回到**个人控制台 > 我的Notebook**内重新启动实例，继续开发。

**Q3：GPU免费额度用完了怎么办？**
后续我们将通过ModelScope社区内运营活动，发放免费算力资源至活动参与者，您可以持续关注我们的社区，以及加入**ModelScope开发者联盟钉钉交流群**，获取最新产品动态。

**Q4：实例启动、关闭过程中报错了怎么办？**
目前Notebook功能仍处于Beta验证阶段，如若遇到报错情况，您可以留意页面内的错误码，加入**ModelScope开发者联盟钉钉交流群**，提供以下信息至群内管理员：

- 用户名
- 错误类型和错误码
- 启动的实例类型：CPU或GPU

提交错误信息后，我们将由专人跟进您的问题并解决，感谢您的反馈和支持。

**Q5：不小心关闭了Notebook页面，实例是否会关闭？**
如果您打开这个实例的时长不超过4小时，您创建的实例不会关闭。您可以回到**个人控制台 > 我的Notebook**点击**查看notebook**按钮，重新打开实例界面。

**Q6：我在Notebook里新下载的依赖包以及保存的文件，关闭实例并再启动后是否还会保留？**
由于官方镜像需要维护，您每次打开实例后，新下载的依赖包仅在此次实例启动过程中有效。您关闭实例并重新启动实例后，您已下载的依赖包将自动清除，需要重新下载。
目前ModelScope Notebook仍处于Beta验证阶段，**您在Notebook内创建的个人文件暂不支持保存**。请您在关闭实例前，手动导出文件至本地，下次启动实例后可上传文件继续编辑。后续版本中，我们将尽快提供云端数据保存功能支持。
